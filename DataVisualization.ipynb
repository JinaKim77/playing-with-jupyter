```python
import pandas as pd  #ways of manipulating data
import numpy as np #numpy is a maths package
import matplotlib.pyplot as plt  #provides the basic plotting functionality that Pandas uses to produce charts and graphs
```

```python
data = pd.read_csv("csv/swineflu.csv")
print(data)
```

```python
print(data['Date'])
print(data['Country'])
```

```python
print(data['Cumulative no. of cases'])
```

```python
#extract Country and Cases from this dataset.
Date = data["Date"]
#Cases = data["Cumulative no. of cases"]
Deaths = data["Cumulative no. of deaths"]

#store these data into two different lists. We need to create two empty lists first
x=[]
y=[]

#use a method list() which converts a dataset into Python list.
x=list(Date)
y=list(Deaths)
```

```python
#print x and y
print(x,y)
```

```python
#Scatter plot from CSV data in Python
plt.scatter(x,y)
plt.xlabel('Date->')
plt.ylabel('Cases->')
plt.title('Data')
plt.show()
```

```python
# Bar Plot from CSV data in Python
plt.bar(x,y)
```
